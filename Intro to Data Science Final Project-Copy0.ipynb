We will use the gradient descent learning algorithm to perform our linear regression. To compare our models we will evaluate the correlation coefficient between the predicted values for each model and our dataset values.

```python
#Computes cost function. For logging purposes
def compute_cost(features, values, theta):
    m = float(len(values))
    sum_of_square_errors = np.square(np.dot(features, theta) - values).sum()
    cost = sum_of_square_errors / (2*m)
    return cost
    
 #gradient descent algorithm, updates iteratively.    
def gradient_descent(features, values, theta, alpha, num_iterations):
    m = float(len(values))
    cost_history = []
    for i in range(num_iterations):
        errors = (np.dot(features, theta) - values)
        theta = theta - (alpha*np.dot(errors,features)*float((1/m)))
        cost_history.append(compute_cost(features, values, theta))
    
    return theta, pd.Series(cost_history)

def predictions(df):
    
    m = len(values)
    
    features['ones'] = np.ones(m)
    features_array = np.array(features)
    values_array = np.array(values).flatten()
    
    alpha = 1
    num_iterations = 100
    
    #initialize
    theta_gradient_descent = np.ones(len(features.columns))
    theta_gradient_descent, cost_history = gradient_descent(features_array, values_array, theta_gradient_descent, alpha, num_iterations)
    predictions = np.dot(features_array, theta_gradient_descent)
    return predictions
    
    
def analysis():
    data = np.array(params_fin['ENTRIESn_hourly'])
    prediction =  np.array(predictions(params_fin))
    params_fin[pred_column] = prediction
    
    r_squared = 1 - (np.square(data - prediction).sum())/((len(data)-1) * np.var(data, axis=None, dtype=None, out=None, ddof=0, keepdims=False))
    
    return r_squared
with open("model_fit.csv", 'w') as master_file:
    master_file.write('feature_set, R^2\n')
    
    for i in range(len(potential_features)):
        features = potential_features[i];
        pred_column = "prediction_%i" %(i)
        r = analysis()
        row = "%s, %f\n" %(feature_names[i], r)
        master_file.write(row)
    
    
fits = pd.read_csv("model_fit.csv")
print fits
```

```python
# -*- coding: utf-8 -*-
import logging
import csv, sys

reader=csv.reader(sys.stdin,delimiter='\t')
entries = 0
days = 0
old_key = None
with open("output.csv", 'w') as master_file:
	master_file.write( '{0},{1}\n'.format('Day', 'Avg_Hourly_Entries'))
	for row in reader:
		data = str(row).strip().split(',')
	
		if len(data) !=2 or data[1][4:-3] == 'ENTRIESn_hourly':

			continue
	
		new_key, value = data
		new_key = new_key[4:-2]
		value = (value[4:-4])

		if old_key != None and new_key !=old_key:
			master_file.write( '{0},{1}\n'.format(old_key, float(entries)/days))
			entries = 0
			days = 0
		
		old_key = new_key
		entries += float(value)
		days += 1
	if old_key != None:
		master_file.write( '{0},{1}\n'.format(old_key, float(entries)/days))

```

### Conclusion:

At this point we can conclude that at least within the month of May traffic patterns were not dominated by weather conditions, and weather conditions alone have weak predictive power of MTA ridership. However, given our limited data set these conclusions are from definitive. Factors such as temperature and precipitation could have very different effects throughout the year with seasonal fluctions. Also there was not enough time to collect large enough sample data to gain a good sense of effects of less common weather events such as fog or thunder. Extreme weather events such as very high winds or high levels rain or snow could also have affects but not enough data was collected to make those assessments. Higher resolution weather data could also be used to see if weather has an affect on the distribution of traffic over a single day.
